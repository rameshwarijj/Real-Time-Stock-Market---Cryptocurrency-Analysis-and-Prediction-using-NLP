Interpretation of Sentiment Scores:

+1.0 to +0.2 → Positive News 📈

-0.2 to -1.0 → Negative News 📉

-0.2 to +0.2 → Neutral News ⚖️

```python
import requests
import pandas as pd
from textblob import TextBlob

#  Replace with your actual GNews API Key
GNEWS_API_KEY = "da87cc30e81b84f3eaa5e5130dd9db00"  
GNEWS_BASE_URL = "https://gnews.io/api/v4/search"

#  Function to fetch real-time cryptocurrency news
def get_crypto_news():
    params = {
        "q": "cryptocurrency OR Bitcoin OR Ethereum OR Binance",
        "token": GNEWS_API_KEY,
        "lang": "en",
        "max": 10  # Fetch latest 10 articles
    }
    response = requests.get(GNEWS_BASE_URL, params=params)
    data = response.json()

    if "articles" in data:
        news_list = [{"title": article["title"], "source": article["source"]["name"], 
                      "published_at": article["publishedAt"], "url": article["url"]}
                     for article in data["articles"]]
        return pd.DataFrame(news_list)
    else:
        print(f" Error fetching news: {data}")
        return pd.DataFrame()

#  Function to perform sentiment analysis
def analyze_sentiment(news_df):
    if news_df.empty:
        return None

    sentiment_scores = []
    for title in news_df["title"]:
        sentiment = TextBlob(title).sentiment.polarity  # -1 (negative) to +1 (positive)
        sentiment_scores.append(sentiment)

    news_df["sentiment"] = sentiment_scores
    return news_df

#  Fetch cryptocurrency news
crypto_news = get_crypto_news()

#  Analyze sentiment
news_sentiment = analyze_sentiment(crypto_news)

#  Display news with sentiment
from IPython.display import display
display(news_sentiment)
```

```python
import requests
import pandas as pd
from textblob import TextBlob

#  Replace with your actual GNews API Key
GNEWS_API_KEY = "da87cc30e81b84f3eaa5e5130dd9db00"  
GNEWS_BASE_URL = "https://gnews.io/api/v4/search"

#  Function to fetch real-time cryptocurrency news
def get_crypto_news():
    params = {
        "q": "cryptocurrency OR Bitcoin OR Ethereum OR Binance",
        "token": GNEWS_API_KEY,
        "lang": "en",
        "max": 10  # Fetch latest 10 articles
    }
    response = requests.get(GNEWS_BASE_URL, params=params)
    data = response.json()

    if "articles" in data:
        news_list = [{"title": article["title"], "source": article["source"]["name"], 
                      "published_at": article["publishedAt"], "url": article["url"]}
                     for article in data["articles"]]
        return pd.DataFrame(news_list)
    else:
        print(f" Error fetching news: {data}")
        return pd.DataFrame()

#  Function to perform sentiment analysis
def analyze_sentiment(news_df):
    if news_df.empty:
        return None

    sentiment_scores = []
    for title in news_df["title"]:
        sentiment = TextBlob(title).sentiment.polarity  # -1 (negative) to +1 (positive)
        sentiment_scores.append(sentiment)

    news_df["sentiment"] = sentiment_scores
    return news_df

#  Fetch cryptocurrency news
crypto_news = get_crypto_news()

#  Analyze sentiment
news_sentiment = analyze_sentiment(crypto_news)

#  Display full news titles with sentiment
if not news_sentiment.empty:
    print("\n📰 **Latest Crypto News with Sentiment Analysis:**\n")
    for index, row in news_sentiment.iterrows():
        sentiment_label = "📈 Positive" if row["sentiment"] > 0 else "📉 Negative" if row["sentiment"] < 0 else "⚖️ Neutral"
        print(f"🔹 **Title:** {row['title']}\n   🏛 **Source:** {row['source']}\n   🕒 **Published At:** {row['published_at']}\n   📊 **Sentiment:** {sentiment_label}\n   🔗 **URL:** {row['url']}\n")
else:
    print(" No news found.")
```
